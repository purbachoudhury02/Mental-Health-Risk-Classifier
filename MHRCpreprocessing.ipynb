```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv("mental_health_risk_dataset (1).csv")
print("Original shape:", df.shape)
print("\nText columns that need encoding:")
print(df.select_dtypes(include='object').columns.tolist())
```

```python
le = LabelEncoder()

text_columns = ['gender', 'marital_status', 'education_level', 'employment_status']

for col in text_columns:
    print(f"\n{col} original values:", df[col].unique())
    df[col] = le.fit_transform(df[col])
    print(f"{col} encoded values:", df[col].unique())
```

```python
print("Data types after encoding:")
print(df.dtypes)
print("\nAny remaining text columns:", df.select_dtypes(include='object').columns.tolist())
```

```python
X = df.drop(columns=['mental_health_risk'])
y = df['mental_health_risk']

print("Features shape:", X.shape)
print("Target shape:", y.shape)
print("\nFeature columns:", X.columns.tolist())
print("\nTarget value counts:")
print(y.value_counts())
```

```python
# Create a composite stress score combining multiple stress indicators
df['composite_stress_score'] = (
    df['work_stress_level'] + 
    df['academic_pressure_level'] + 
    df['financial_stress_level']
) / 3

# Create a wellbeing score
df['wellbeing_score'] = (
    df['sleep_hours'] + 
    df['physical_activity_hours_per_week'] + 
    df['social_support_score'] + 
    df['job_satisfaction_score']
) / 4

print("New features added:")
print(df[['composite_stress_score', 'wellbeing_score']].describe())
```

```python
X = df.drop(columns=['mental_health_risk'])
y = df['mental_health_risk']

# Save processed data
df.to_csv('processed_data.csv', index=False)
print("Processed data saved.")
print("Final shape:", df.shape)
```
