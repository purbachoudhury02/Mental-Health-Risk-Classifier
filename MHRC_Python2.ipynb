```python
def analysis():
    Sleep_Hours=int(input("Enter your total number of sleep hours: "))
    Stress_level=int(input("On a scale of 1-10 , How stressed are you?"))
    if(Sleep_Hours <=6 and Stress_level>=6):
        return "At risk"
    elif(Sleep_Hours <=6 and Stress_level<=5):
        return "Medium risk,Sleep more!"
    elif(Sleep_Hours >=7  and Stress_level>=6):
        return "Medium risk,Stress less!"
    else:
        return "Not at risk"
result=analysis()
print(result)
    
```
