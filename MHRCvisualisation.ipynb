```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("mental_health_risk_dataset (1).csv")
sns.set_style("whitegrid")
```

```python
plt.figure(figsize=(8, 5))
risk_labels = {0: 'Low Risk', 1: 'Medium Risk', 2: 'High Risk'}
df['mental_health_risk'].map(risk_labels).value_counts().plot(kind='bar', color=['green', 'orange', 'red'])
plt.title('Mental Health Risk Distribution')
plt.xlabel('Risk Level')
plt.ylabel('Number of People')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

The majority of individuals in the dataset belong to the Medium Risk category, while the High Risk category has the fewest samples, indicating a moderate class imbalance in the target variable.

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='mental_health_risk', y='sleep_hours', data=df, palette=['green', 'orange', 'red'])
plt.title('Sleep Hours by Mental Health Risk Level')
plt.xlabel('Risk Level (0=Low, 1=Medium, 2=High)')
plt.ylabel('Sleep Hours')
plt.show()
```

The boxplot suggests that individuals with higher mental health risk tend to sleep fewer hours on average. This indicates a possible relationship between reduced sleep and increased mental health risk.

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='mental_health_risk', y='work_stress_level', data=df, palette=['green', 'orange', 'red'])
plt.title('Work Stress Level by Mental Health Risk')
plt.xlabel('Risk Level (0=Low, 1=Medium, 2=High)')
plt.ylabel('Work Stress Level')
plt.show()
```

Individuals in higher mental health risk categories tend to report higher work stress levels. This suggests a positive relationship between work stress and mental health risk.

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='mental_health_risk', y='screen_time_hours_per_day', data=df, palette=['green', 'orange', 'red'])
plt.title('Screen Time by Mental Health Risk')
plt.show()
```

The screen time distributions are very similar across all mental health risk categories. This suggests that screen time alone may not be a strong indicator of mental health risk in this dataset.

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='gender', hue='mental_health_risk', data=df, palette=['green', 'orange', 'red'])
plt.title('Gender vs Mental Health Risk')
plt.show()
```

The distribution of mental health risk levels appears similar across genders, suggesting that gender alone may not be a strong predictor of mental health risk in this dataset.

```python
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
plt.figure(figsize=(16, 12))
sns.heatmap(df[numerical_cols].corr(), annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Correlation Between All Numerical Features')
plt.tight_layout()
plt.show()
```

The features most correlated with mental_health_risk are panic_attack_history (Positive correlation of 0.42),anxiety_score or depression_score (Both have an identical positive correlation of 0.32),sleep_hours (Negative correlation of -0.31)
